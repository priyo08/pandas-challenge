```python
# set a = [1,2,3,4,]

# set b = [2,3,5,6]

# inner join a, b  = {2,3} only common elements
# left join => {1,2,3,4} left  elements + common  
# left outer join => {1,4} only present left but not in common ele.
# full outer join => {1,2,3,4,5,6} left + common + right.
```

```python
s_data = pd.DataFrame({"s_id": [1,2,3,4],"sch_name":['a', 'b', 'b', 'c'], "st_name": ['1a', '2a', '3a', '4a']  })
s_data
```

```python
sch_data = pd.DataFrame({"sch_id": [11,22,33,44],"sch_name":['a', 'b', 'd', 'e'], "princ_name": ['a1a', 'a2a', 'a3a', 'a4a']  })
sch_data
```

```python
s_data.merge(sch_data, how='inner', on='sch_name') # inner join
```

```python
s_data.merge(sch_data, how='left', on='sch_name') # left join
```

```python
s_data.merge(sch_data, how="outer", on="sch_name", indicator=True).query('_merge=="left_only"') # left outer join
```

```python
s_data.merge(sch_data, how="outer", on="sch_name", indicator=True) #Full outer join
```

```python
s_data.merge(sch_data, how='right', on='sch_name') # Right join
```

```python
s_data.merge(sch_data, how='right', on='sch_name', indicator=True).query('_merge=="right_only"') # Right outer.
```
